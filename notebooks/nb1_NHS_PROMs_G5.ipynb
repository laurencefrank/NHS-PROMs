**Example of machine learning analyses on these data**
Huber et. al. (2019) used several machine learning classifiers to predict patient-reported outcomes following hip and knee replacement surgery. They conclude that machine learning improved the prediction of the NHS-data, compared to (simle) linear models. The NHS uses multiple linear regression to predict patient-reported outcomes.

However there is a **methodological issue** in Huber's paper. The authors use the minimal important difference measure (MID) to determine whether a patient has improved or not after surgery. The patient-reported outcome was measured with the visual analog scale (VAS) and with the Q-score. The Q-score is the sum of the Oxford Hip Score (OHS) or the sum of the Oxford Knee Score (OKS).

The outcome for VAS for example, was calculated as follows:

Y_VAS = 0 when post-operative VAS - pre-operative VAS <= 0.5 standard deviation

Y_VAS = 1 when post-operative VAS - pre-operative VAS > 0.5 standard deviation

The methodological issues arises when Huber et.al. use the pre-operative VAS score as a predictor in the machine learning models. The pre-operative VAS score has been used to calculate the outcome `y` and therefore the outcome 'y' contains information about the pre-operative status of the patients. It is then no surprise that authors report that pre-operative VAS is the best predictor of post-operative outcome! It is very likely that the predictive properties of the machine learning models are overestimated in this paper. 

To get a better estimate of the predictive properties of the model there two options:
1. When the outcome variable 'y' is calculated based on the difference between pre-op scores and post-op scores, do not use the same pre-op scores as a predictor in your models.
2. Use unmodified post-op scores as outcome variable and pre-op scores as predictor.
